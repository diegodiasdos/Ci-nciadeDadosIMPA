Note que nos dados provenientes da tabela municipios.csv o nome dos municípios estão todos acentuados, no entanto na tabela amazonas os nomes estão sem acento. Logo é necessario remover os acentos dos nomes da tabela municipios.csv.

```python
# -*- coding: latin-1 -*-
# -*- coding: UTF-8 -*-

# remover acentos 
import unidecode
import csv
import pandas as pd
import sys 


def remove_acentos(feed):
  # abrir o arquivo
  csv_f = open(feed, encoding="latin-1", mode='r')
  #ler o arquivo 
  csv_str = csv_f.read()
  csv_str_remove_acento = unidecode.unidecode(csv_str)
  csv_f.close()
  csv_f = open(feed, 'w')
  csv_f.write(csv_str_remove_acento)
  return True

if __name__ == "__main__":
    
    remove_acentos('municipios.csv')
    print(csv_f)
    
    
# Verificar se está certo 

    
```
